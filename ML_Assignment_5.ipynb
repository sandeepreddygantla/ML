```python
import pandas as pd
from sklearn.naive_bayes import CategoricalNB
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_excel('CoffeeShop.xlsx')
df
```

```python
# Perform label encoding
label_encoder = LabelEncoder()
df_encoded = df.copy()
df_encoded['Location'] = label_encoder.fit_transform(df_encoded['Location'])
df_encoded['Product'] = label_encoder.fit_transform(df_encoded['Product'])
df_encoded['Feedback'] = label_encoder.fit_transform(df_encoded['Feedback'])
df_encoded
```

```python
# Initialize the accuracy list
accuracies = []

# Perform leave-one-out cross-validation
for i in range(len(df_encoded)):
    # Split the data into training and testing sets
    X_train = df_encoded.drop(index=i)
    y_train = df_encoded['Feedback'].drop(index=i)
    X_test = df_encoded.iloc[i:i+1]
    y_test = df_encoded['Feedback'].iloc[i:i+1]

    # Create and train the Categorical Naive Bayes classifier
    clf = CategoricalNB()
    clf.fit(X_train, y_train)

    # Make predictions on the test set
    y_pred = clf.predict(X_test)

    # Compute accuracy and append to the list
    accuracy = accuracy_score(y_test, y_pred)
    accuracies.append(accuracy)

accuracies
```

```python
# Compute the average accuracy
average_accuracy = sum(accuracies) / len(accuracies)
print("Average Accuracy:", average_accuracy)
```
